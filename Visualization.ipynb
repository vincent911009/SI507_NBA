```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stats = pd.read_csv("Combined_stats.csv")
```

Choose one stat that you want to compare and remain the numbers of playersyou want to plot.

```python
x = input("Type what stats you want to compare: ")
y = int(input("Type how many people you want to remain: "))
highest_scoring = stats[stats["G_x"] > 70].sort_values(x, ascending=False).head(y)
highest_scoring.plot.bar("Player",x)
plt.ylabel(x)
```

Consider more stats: First, choose one stat that you want to compare and remain numbers of players. Finally, choose the final stats you want to compare and remain the final numbers of players you want to plot.

```python
x = input("Type what stats you want to compare: ")
y = int(input("Type how many people you want to remain: "))
z = input("Type what stats you also want to compare: ")
m = int(input("Type how many people you want to remain at last: "))
highest_scoring = stats[stats["G_x"] > 70].sort_values(x, ascending=False).head(y)
highest_scoring = highest_scoring.sort_values(z, ascending=False).head(m)
highest_scoring.plot.bar("Player",z)
plt.ylabel(x)
```

Choose one stats you want to compare and plot the highest stats in each year.

```python
y = input("Choose one stat you want to compare for each year:")
highest_scoring_season = stats.groupby("Year").apply(lambda x: x.sort_values(y, ascending=False).head(1))
highest_scoring_season.plot.bar("Year",y)
```
